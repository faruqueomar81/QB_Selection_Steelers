```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
qb_stats=pd.read_excel(r'C:\Users\faruq\Documents\College Football Player Stats.xlsx',sheet_name="QB_STATS_2021")
qb_stats_2020=pd.read_excel(r'C:\Users\faruq\Documents\College Football Player Stats.xlsx',sheet_name="QB_STATS_2020")
qb_stats.head()
```

```python
qb_stats.head()
qb_stats.shape
```

```python
qb_stats_2020.shape
```

```python
qb_stats['Indicator']='2021'
qb_stats_2020['Indicator']='2020'
qb_stats.head()
```

```python
qb_stats_2020.head()
```

```python
qb_stats=qb_stats.append(qb_stats_2020, ignore_index=True) #adding rows from the 2020 stats to the original table
qb_stats.shape
```

```python
qb_stats=qb_stats.rename(columns={'U':'University_Name','Player':'QB_Name','G':'Games_Played','Rk':'Depth Chart'}) #renaming two columns for better clarity
qb_stats.head()
```

```python
qb_stats[['QB_Name','Reject']]=qb_stats['QB_Name'].str.split('\\',expand=True) #or you can use raw string str.split(r'\',expand=True)
qb_stats.head()
```

```python
qb_stats.drop(['Reject'],axis=1,inplace=True)
qb_stats.head()
```

```python
qb_stats['Pass_Completion_Percent']=round((qb_stats['Passing Cmp']/qb_stats['Passing Att'])*100,2)
qb_stats['Passer_Rating']=round(((qb_stats['Passing Yds']*8.4+qb_stats['Passing TD']*330-qb_stats['Passing Int']*205)/qb_stats['Passing Att']),2)
qb_stats.head()
```

```python
qb_stats.columns.get_loc('Pass_Completion_Percent')#If you table is really wide, just manually counting is not realistic
```

```python
qb_stats.columns.get_loc('Passing Yds')
```

```python
column_titles=list(qb_stats.columns)
new_cols = column_titles[:6] + [column_titles[-2]] + column_titles[6:-2]+column_titles[-1:] #list indexing. -1 because its from behind
qb_stats = qb_stats[new_cols]
qb_stats.head()
```

```python
rushing_stats=pd.read_excel(r'C:\Users\faruq\Documents\College Football Player Stats.xlsx',sheet_name="RUSHING_STATS")
rushing_stats.head()
```

```python
rushing_stats=rushing_stats.iloc[:,0:7] #filtered out the columns that are not needed.iloc(row,column)
rushing_stats
```

```python
rushing_stats=rushing_stats.rename(columns={'G':'Games Played','Player':'Player_Name'}) #renaming two columns for better clarity
rushing_stats.head()
```

```python
rushing_stats[['Player_Name','Reject']]=rushing_stats['Player_Name'].str.split('\\',expand=True) #or you can use raw string str.split(r'\',expand=True)
rushing_stats.head()
```

```python
rushing_stats.drop('Reject', axis=1,inplace=True)
rushing_stats.head()
```

```python
qb_stats=pd.merge(qb_stats,rushing_stats,left_on='QB_Name',right_on='Player_Name',how='left')
qb_stats.head()
```

```python
qb_stats.dropna(how='all',inplace=True) # all lets you take out rows with all null value. and inplace=True help you change the dataframe permanantly
qb_stats.head()
```

```python
%matplotlib inline
plt.style.use('seaborn')
qb_stats.plot.bar(x='QB_Name',y='Passer_Rating',rot=90,figsize=(15,10),color='green')
#qb_stats['Rushing Yds'].plot.bar(x='QB_Name',y='Rushing Yds',secondary_y=True,rot=90,figsize=(15,10),color='red');
plt.xlabel('QB_Name')
plt.ylabel('Passer_Rating')
plt.title('2020 College Football Quarterback Passer Rating')
plt.show()
```
